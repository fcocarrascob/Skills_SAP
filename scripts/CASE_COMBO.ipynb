# Selección de Load Cases / Load Combinations para Output
Conecta a SAP2000, lista los casos o combinaciones disponibles y los selecciona para output.

```python
import comtypes.client

# ── Conectar a instancia de SAP2000 activa ────────────────────────────
try:
    sap_object = comtypes.client.GetActiveObject("CSI.SAP2000.API.SapObject")
    SapModel   = sap_object.SapModel
    version    = str(sap_object.GetOAPIVersionNumber())
    model_path = str(SapModel.GetModelFilename())
    print(f"Conectado a SAP2000 v{version}")
    print(f"Modelo: {model_path or '(sin guardar)'}")
except Exception as e:
    print(f"ERROR: No se pudo conectar a SAP2000 — {e}")
    raise
```

```python
# ── Selección de Load Cases o Load Combinations para Output ──────────

# 1. Elegir tipo
print("¿Qué desea seleccionar para output?")
print("  1 — Load Cases")
print("  2 — Load Combinations")
tipo = input("Ingrese opción (1 o 2): ").strip()

if tipo not in ("1", "2"):
    raise ValueError("Opción inválida. Ingrese 1 o 2.")

# 2. Obtener lista completa
if tipo == "1":
    raw = SapModel.LoadCases.GetNameList(0, [])
    nombres = list(raw[1])
    etiqueta = "Load Cases"
else:
    raw = SapModel.RespCombo.GetNameList(0, [])
    nombres = list(raw[1])
    etiqueta = "Load Combinations"

if not nombres:
    print(f"No hay {etiqueta} definidos en el modelo.")
else:
    # 3. Deseleccionar todo y seleccionar todos
    ret = SapModel.Results.Setup.DeselectAllCasesAndCombosForOutput()
    assert ret == 0, f"DeselectAll falló: {ret}"

    for nombre in nombres:
        if tipo == "1":
            ret = SapModel.Results.Setup.SetCaseSelectedForOutput(nombre, True)
            assert ret == 0, f"SetCaseSelectedForOutput('{nombre}') falló: {ret}"
        else:
            ret = SapModel.Results.Setup.SetComboSelectedForOutput(nombre, True)
            assert ret == 0, f"SetComboSelectedForOutput('{nombre}') falló: {ret}"

    print(f"\n✓ Todos los {etiqueta} seleccionados para output ({len(nombres)}):")
    for nombre in nombres:
        print(f"  • {nombre}")
```
